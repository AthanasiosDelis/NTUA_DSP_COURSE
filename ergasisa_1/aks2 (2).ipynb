```python
import numpy as np
import scipy as sp
import librosa
import matplotlib.pyplot as plt
import pywt
import sounddevice as sd
from scipy import signal
```

```python
t = np.linspace(0,2,1000)
x1 = 2*np.cos(2*np.pi*70*t) + 3*np.sin(2*np.pi*140*t) + 0.15*np.random.normal()
x2 = 1.7*np.cos(2*np.pi*90*t) + 0.15*np.random.normal() + 1.7*signal.unit_impulse(1000,625) + signal.unit_impulse(1000,800)
plt.plot(t,x1)
plt.show()

f1,t1,x1s = signal.stft(x1,fs = 1000.0,nperseg=40,noverlap = 20)

plt.pcolormesh(t1,f1,np.abs(x1s))
#plt.matshow(x1)
```

```python
x2 = 1.7*np.cos(2*np.pi*90*t) + 0.15*np.random.normal() + 1.7*signal.unit_impulse(1000,625) + signal.unit_impulse(1000,800)
f2,t2,x2s = signal.stft(x2,fs = 1000.0,nperseg=40,noverlap = 20)
plt.pcolormesh(t2,f2,np.abs(x2s))
print(f2,"-------",t2,"--------",x2s)
```

```python
t = np.linspace(0,2,1000,endpoint = False)
x1 = 2*np.cos(2*np.pi*70*t) + 3*np.sin(2*np.pi*140*t) + 0.15*np.random.normal()
s = np.linspace(1,6,20)
coefs,freqs = pywt.cwt(x1,s,'cmor3.0-1.0')
print(coefs.shape)
%matplotlib inline
f = freqs*1000
plt.pcolormesh(t,f,np.abs(coefs))
plt.xlabel('Time (sec)')
plt.ylabel('Frequency(Hz)')
plt.title('CWT of the signal segment')
```
