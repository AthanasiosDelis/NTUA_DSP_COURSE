```python
#!/usr/bin/env python
from __future__ import division 
import sys
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import pywt
import statistics


def mean(sig):
    l=statistics.mean(sig)
    return l
def std(sig):
    l=np.std(sig)
    return l
def ste(signal,window=20,fs=20):
    samp=window*fs
    Nsamp=int(samp)
    h=np.hamming(Nsamp)
    STE=np.convolve(signal*signal,h)
    return STE

zfile=np.load(r"C:\Users\Thanasis\Documents\DSP\DSP20_LAB2\sleep_03.npz")
xacc=np.zeros(len(zfile['acc']))
yacc=np.zeros(len(zfile['acc']))
zacc=np.zeros(len(zfile['acc']))
xgyr=np.zeros(len(zfile['gyr']))
ygyr=np.zeros(len(zfile['gyr']))
zgyr=np.zeros(len(zfile['gyr']))
hrm=np.zeros(len(zfile['hrm']))

w=len(zfile['acc'])-11000
for i in range(w):
    xacc[i]=zfile['acc'][i][0]
    yacc[i]=zfile['acc'][i][1]
    zacc[i]=zfile['acc'][i][2]


e=len(zfile['gyr'])-11000
for i in range(e):
    xgyr[i]=zfile['gyr'][i][0]
    ygyr[i]=zfile['gyr'][i][1]
    zgyr[i]=zfile['gyr'][i][2]
s=len(zfile['hrm'])
for i in range(s):
    hrm[i]=zfile['hrm'][i]

xo=[]

x=ste(xacc)
y=ste(yacc)
z=ste(zacc)
x1=ste(xacc)
y1=ste(yacc)
z1=ste(zacc)
x2=ste(xacc)


xo.append(x)
xo.append(y)
xo.append(z)
xo.append(x1)
xo.append(y1)
xo.append(z1)
xo.append(x2)
yo=["xacc","yacc","zacc","xgyr","ygyr","zgyr","hrm"]

k=0
array14=np.zeros((7,5))
print('Mean Min Max Standar deviation:')
for i in xo:
    print('\n')
    array14[k][0]=mean(i)
    array14[k][1]=max(i)
    array14[k][2]=min(i)
    array14[k][3]=std(i)
    print(yo[k],': ',*(array14[k][j] for j in range(4)) ,sep=' ')           
    k+=1
zfile=np.load(r"C:\Users\Thanasis\Documents\DSP\DSP20_LAB2\step_03.npz")
```
