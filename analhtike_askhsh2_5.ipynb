```python
import numpy as np
import scipy as sp
import librosa 
import matplotlib.pyplot as plt
import cmath as cm
```

```python
w_fi=np.linspace(-2*np.pi,2*np.pi,2000)
W_fi=np.linspace(-np.pi,np.pi,1000)
T = 1
j=cm.sqrt(-1)
s_la=j*W_fi
z=np.exp(j*w_fi)
Ha_fi=5/(s_la**2+2*s_la+5)
H_fi=2.5*np.exp(-1)*z**(-1)*np.sin(2)/(1-2*np.exp(-1)*z**(-1)*np.cos(2)+np.exp(-2)*z)
plt.plot(w_fi,H_fi)
plt.title("|H(exp(j*w))|")
plt.show()
plt.plot(W_fi,Ha_fi)
plt.title("|Ha(jw)|")
plt.show()
```
