```python
#1.2
def ste(x, window='hamming', window_length=400):
    if isinstance(window, str):#npt nessesarilly needs, it narrows the possible windows to be created
        window = sp.signal.get_window(window, window_length)#garuntees to make the window string, no tuple 
    
    return sp.signal.convolve(x**2, window, mode="same")#type length of max(x,window) in samples
```

```python
fs=20 #sampling freq acc,gyr
fsh=5 #sampling freq hrm

def plotter(signal,fs,name="STE signal",window='hamming',time=20):
    stesig=ste(signal,window, time*fs)
    #
    sec=np.linspace(0,len(stesig),len(stesig))/fs
    plt.plot(sec,stesig)
    p=str(name)
    plt.title(p)
    plt.xlabel("Sec")
    plt.show()
    return sec,stesig
plotter(xacc,fs)
def plotterbulk(signal,freq,title="Signal and its STE"):
    #
    sec,stesig=plotter(signal,fs=freq,name="STE signal",window='hamming',time=20)
    fig, ax7 = plt.subplots()

    ax7.set_xlabel('Sec')
    ax7.set_ylabel('Input signal', color='red')
    ax7.plot(sec, signal, color='red')
    ax7.tick_params(axis='y')
    #
    ax8 = ax7.twinx()  # instantiate a second axes that shares the same x-axis

    ax8.set_ylabel('ste', color='blue')  # we already handled the x-label with ax1
    ax8.plot(sec, stesig, color='blue')
    ax8.tick_params(axis='y')

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    p=str(title)
    plt.title(p)
    plt.show()
plotterbulk(xacc,fs)
plotterbulk(xgyr,fs)
plotterbulk(hrm,fsh)



"""
stexgyr=ste(xgyr,'hamming', 400)
stexhrm=ste(hrm,'hamming', 100)    
sech=np.linspace(0,len(stexhrm),len(stexhrm))/fsh
plt.plot(sec,stexacc)
plt.title("STE xgyr")
plt.xlabel("Sec")
plt.show()
plt.plot(sech,stexhrm)
plt.title("STE xhrm")
plt.xlabel("Sec")
plt.show()
"""
```

```python
from __future__ import division
#1.3
def plotterbeauty(signal,fs,signame,time=20):
    hl=int((time/2)*fs)
    stftacc=librosa.stft(signal,n_fft=1024,hop_length=hl,win_length=int(fs*time))#stft window length 40 samples overlappng 20 samples 50%
    #
    sec=np.linspace(0,int(len(signal)/fs),round(len(signal)/(fs*time/2)))
    #
    freq= np.linspace(0,10,np.shape(stftacc)[0])
    
    plt.pcolormesh(sec,freq,np.log(np.abs(stftacc)))
    python_power=str(signame)
    plt.title('spectrogram of log('+python_power+') stft ')
    plt.xlabel('Sec')
    plt.ylabel('Freq')
    plt.show()
plotterbeauty(xacc,fs,"x[n]")
plotterbeauty(hrm,fsh,"hrm[n]")
```
